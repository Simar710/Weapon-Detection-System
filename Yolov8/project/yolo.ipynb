# So We have csv file with labels of all the images we have now creating labels file for individual images

```python
import pandas as pd
import os
from os import path

# Define the base path for images and labels
base_path = 'C:/Users/manjo/4780/project/full_dataset'
images_path = path.join(base_path, 'images')
labels_path = path.join(base_path, 'labels')

# Make sure the labels directory exists
if not path.exists(labels_path):
    os.makedirs(labels_path)
    
# Load the CSV file
csv_file = os.path.join(base_path, 'labels', 'labels.csv').replace('\\', '/')
df = pd.read_csv(csv_file)


# Iterate over the dataframe and handle rows where 'filename' is not a string
for _, row in df.iterrows():
    if isinstance(row['filename'], str):
        # Get image dimensions and filename without extension
        img_width, img_height = row['width'], row['height']
        filename_no_ext = os.path.splitext(row['filename'])[0]
        
        # Create label file path
        label_file = os.path.join(labels_path, f"{filename_no_ext}.txt")
        
        # Write placeholder content to the label file
        with open(label_file, 'w') as file:
            file.write(f"{row['class']} 0.5 0.5 1 1\n")  # Placeholder values for the bounding box
    else:
        print(f"Invalid filename at index {_}: {row['filename']}")
        
#  Print a message to indicate that label files have been created
print(f"Label files created at: {labels_path}")
```

# spliting the data into training and validation because with labels because yolo follows a structure where it needs trainging and validation images with labels of that image with it.

```python
import os
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split

# Paths
images_folder = 'C:/Users/manjo/4780/project/full_dataset/images'
labels_folder = 'C:/Users/manjo/4780/project/full_dataset/labels'  # If labels are in the same folder, this can be the same as images_folder
spreadsheet_path = 'C:/Users/manjo/4780/project/full_dataset/labels/labels.csv'

# Read the spreadsheet
data = pd.read_csv(spreadsheet_path)

# Convert filenames to strings
data['filename'] = data['filename'].astype(str)

# Split dataset
train_data, val_data = train_test_split(data, test_size=0.2, random_state=42)

# Create directories
train_img_dir = os.path.join(images_folder, 'train')
val_img_dir = os.path.join(images_folder, 'val')
train_lbl_dir = os.path.join(labels_folder, 'train')
val_lbl_dir = os.path.join(labels_folder, 'val')
os.makedirs(train_img_dir, exist_ok=True)
os.makedirs(val_img_dir, exist_ok=True)
os.makedirs(train_lbl_dir, exist_ok=True)
os.makedirs(val_lbl_dir, exist_ok=True)

# Move function
def move_files(df, src_folder, dst_folder, file_ext):
    for filename in df['filename']:
        base_filename = os.path.splitext(filename)[0]
        full_filename = f"{base_filename}{file_ext}"
        src_path = os.path.join(src_folder, full_filename)
        dst_path = os.path.join(dst_folder, full_filename)
        if os.path.exists(src_path):
            shutil.move(src_path, dst_path)
        else:
            print(f"File not found and skipped: {src_path}")

# Move images and labels
move_files(train_data, images_folder, train_img_dir, '.jpg')
move_files(val_data, images_folder, val_img_dir, '.jpg')
move_files(train_data, labels_folder, train_lbl_dir, '.txt')
move_files(val_data, labels_folder, val_lbl_dir, '.txt')

print("Images and labels have been successfully organized into train and val folders.")
```

# Converting the classes names from string to integer as it is one of the requirements in yolo to make model understand the class that image belongs to

```python
import os

class_names = ['rifle', 'knife', 'handgun', 'ax', 'sniper', 'pistol', 'shotgun', 'spear', 'eto', 'cutter', 'cleaver', 'explosive']
class_map = {name: idx for idx, name in enumerate(class_names)}

label_dir = r'C:\Users\manjo\4780\project\full_dataset\train\labels'  # Adjust this path according to location of labels file in  train

for label_file in os.listdir(label_dir):
    file_path = os.path.join(label_dir, label_file)
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    with open(file_path, 'w') as file:
        for line in lines:
            parts = line.strip().split()
            if parts[0] in class_map:
                parts[0] = str(class_map[parts[0]])
                file.write(' '.join(parts) + '\n')
```

# checking with different images and finding the coordinates for better understanding

```python
from ultralytics import YOLO

# Load a model
model = YOLO(r"C:/Users/manjo/4780/project/runs/detect/train7/weights/best.pt")


results = model.predict("gun.jpg")
result = results[0]

for box in result.boxes:
  class_id = result.names[box.cls[0].item()]
  cords = box.xyxy[0].tolist()
  cords = [round(x) for x in cords]
  conf = round(box.conf[0].item(), 2)
  print("Object type:", class_id)
  print("Coordinates:", cords)
  print("Probability:", conf)
  print("---")
    
print()
```

# The image will pop up for visuals and better understanding

```python
from ultralytics import YOLO
import cv2

# Load the model
model = YOLO(r"C:/Users/manjo/4780/project/runs/detect/train7/weights/best.pt")

# Predict with the model
results = model('gun.jpg')

# Load the image using OpenCV
image = cv2.imread('gun.jpg')

# Process the results
for result in results:
    boxes = result.boxes
    names = result.names

    # Iterate over the detections
    for box in boxes:
        # Extracting bounding box coordinates, confidence, and class ID
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
        conf = box.conf[0].item()  # Assuming single detection per box
        cls_id = int(box.cls[0].item())

        label = f'{names[cls_id]} {conf:.2f}'
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

# Display the image
cv2.imshow('Detected Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
